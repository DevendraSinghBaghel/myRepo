```python
def checkSanity(chunk,end_ind):
    
    clean_chunk = ''
    start_ind_slice = 0
    end_ind_slice = len(chunk)
    max_slice_size = 7
    for i in range(1000):
        clean_chunk = chunk[start_ind_slice:end_ind_slice]
        index_list = []
        #temp_index=end_ind_slice
        last_7 = clean_chunk[max_slice_size:]
        print (last_7)
        index_list= [ last_7.find(x) for x in ['S','T','Y'] if last_7.find(x) != -1 ]
        print(index_list)
        if len(index_list) == 0:
            break
            
        temp_val = max_slice_size - min(index_list)
        end_ind_slice = end_ind_slice - temp_val
        end_ind       = end_ind - temp_val
    
    return(clean_chunk,end_ind)
        
        
       
        
    
```
